```python
import numpy as np
from google.colab import autoviz

def time_series_multiline(df, timelike_colname, value_colname, series_colname, figscale=1, mpl_palette_name='Dark2'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  figsize = (10 * figscale, 5.2 * figscale)
  palette = list(sns.palettes.mpl_palette(mpl_palette_name))
  def _plot_series(series, series_name, series_index=0):
    if value_colname == 'count()':
      counted = (series[timelike_colname]
                 .value_counts()
                 .reset_index(name='counts')
                 .rename({'index': timelike_colname}, axis=1)
                 .sort_values(timelike_colname, ascending=True))
      xs = counted[timelike_colname]
      ys = counted['counts']
    else:
      xs = series[timelike_colname]
      ys = series[value_colname]
    plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

  fig, ax = plt.subplots(figsize=figsize, layout='constrained')
  df = df.sort_values(timelike_colname, ascending=True)
  if series_colname:
    for i, (series_name, series) in enumerate(df.groupby(series_colname)):
      _plot_series(series, series_name, i)
    fig.legend(title=series_colname, bbox_to_anchor=(1, 1), loc='upper left')
  else:
    _plot_series(df, '')
  sns.despine(fig=fig, ax=ax)
  plt.xlabel(timelike_colname)
  plt.ylabel(value_colname)
  return autoviz.MplChart.from_current_mpl_state()

chart = time_series_multiline(aggregated_data, *['Arrival Time', 'Days', None], **{})
chart
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data2' is your DataFrame

# Convert 'Arrival Time' to datetime if it's not already
data2['Arrival Time'] = pd.to_datetime(data2['Arrival Time'])

# Calculate the total number of people (adults + babies) for each booking
data2['Total People'] = data2['Number of adults'] + data2['Number of babies']

# Group data by weekends and weekdays and calculate the mean 'Paid Ammount'
data2['Day of the Week'] = data2['Arrival Time'].dt.day_name()
data2['Weekend/Weekday'] = data2['Day of the Week'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')

# Group by 'Weekend/Weekday', 'Total People', and 'Arrival Time', and calculate the mean 'Paid Ammount'
aggregated_data = data2.groupby(['Weekend/Weekday', 'Total People', pd.Grouper(key='Arrival Time', freq='W')]).agg({
    'Paid Ammount': 'mean'
}).reset_index()

# Create a line plot to compare the trend of prices for adults and babies based on weekends and weekdays over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=aggregated_data, x='Arrival Time', y='Paid Ammount', hue='Total People', style='Weekend/Weekday', markers=True, palette='tab10')
plt.title('Trend of Prices for Adults and Babies Based on Weekends and Weekdays Over Time')
plt.xlabel('Time')
plt.ylabel('Average Paid Amount')
plt.xticks(rotation=45)
plt.legend(title='Total People', loc='upper right')
plt.show()
```

```python
import pandas as pd

# Assuming 'data2' is your DataFrame

# Convert 'Arrival Time' to datetime if it's not already
data2['Arrival Time'] = pd.to_datetime(data2['Arrival Time'])

# Define the pricing model based on the number of adults and children
pricing_model = {
    (1, 0): 35,
    (2, 0): 70,
    (3, 0): 100,
    (4, 0): 120,
    (1, 1): 55,
    (1, 2): 75,
    (1, 3): 95,
    (2, 1): 90,
    (2, 2): 110,
    (3, 1): 120,
    (3, 2): 145,
    (4, 1): 150,
    (0, 1): 25  # Assuming 1 Child with 0 adults
}

# Function to calculate Proposed Paid Amount based on the number of adults and children
def calculate_proposed_paid_amount(row):
    if row['Arrival Time'].weekday() < 5:  # Check if it's a weekday (0 to 4 represent Monday to Friday)
        key = (row['Number of adults'], row['Number of babies'])
        proposed_paid_amount = pricing_model.get(key, 0)  # Return 0 if the combination is not in the model
        # Apply the 1/3 increment for proposed paid amount on weekdays
        return proposed_paid_amount + (proposed_paid_amount / 3)
    else:
        return row['Paid Ammount']  # Use the existing 'Paid Ammount' on weekends

# Create a new column 'Proposed Paid Amount' using the pricing model for weekdays
data2['Proposed Paid Amount'] = data2.apply(calculate_proposed_paid_amount, axis=1)

data2
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data2' is your DataFrame with 'Arrival Time', 'Paid Ammount', and 'Proposed Paid Amount'

# Convert 'Arrival Time' to datetime if it's not already
data2['Arrival Time'] = pd.to_datetime(data2['Arrival Time'])

# Group the data by week and calculate the mean for 'Paid Ammount' and 'Proposed Paid Amount'
weekly_data = data2.resample('W', on='Arrival Time').mean()

# Create a time series line chart
plt.figure(figsize=(12, 6))
plt.plot(weekly_data.index, weekly_data['Paid Ammount'], label='Current Revenue Trend', linestyle='-', marker='o')
plt.plot(weekly_data.index, weekly_data['Proposed Paid Amount'], label='Proposed Revenue Trend', linestyle='-', marker='o')

plt.title('Weekly Price Trend')
plt.xlabel('Time')
plt.ylabel('Revenue (Average)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```
